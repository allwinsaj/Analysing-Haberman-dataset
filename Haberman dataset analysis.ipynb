# Haberman's Survival Dataset

```python
import pandas as pd                         
import numpy as np                           
import matplotlib.pyplot as plt
import seaborn as sns  
column_names=["Age","operation_years","Lymph_nodes","survivor_status"]  
haberman=pd.read_csv("haberman.csv",names=column_names)
print(haberman)

#First fourline of code I am just import the packages that we want to use in this assignment and also i named it
#my data set doesnt have column names or labels or variables so i assigned the name(names=column_names).

```

```python
haberman["survivor_status"]=haberman["survivor_status"].map({1:"sur_abv_5",2:"dieblw_5"})
print(haberman)
#Here I assign that 1="sur_abv_5",2="dieblw_5"  in class attributes("survivor_status")
```

```python
haberman.shape
#to know the shape of the dataset(rows,columns)
```

```python
haberman_surviveabove5yrs=haberman.loc[haberman["survivor_status"]=="sur_abv_5"]
haberman_diewithin5yrs=haberman.loc[haberman["survivor_status"]=="dieblw_5"]
#Here I am just save the data of who all are surviving more than 5years in "haberman_surviveabove5yrs" 
#And also I am saving the data of who all are died within 5years in "haberman_diewithin5yrs"



```

```python
print("Mean of Age who are alive",np.mean(haberman_surviveabove5yrs["Age"]))#Mean value of Age who are survive more than 5years 
print("Mean of Age who are dead",np.mean(haberman_diewithin5yrs["Age"]))#Mean value of Age who were died within 5years
print("Mean of Age ie:overall(including alive and dead)",np.mean(haberman["Age"]))#Mean value of Age in haberman dataset
print("Mean of operation years who are alive", np.mean(haberman_surviveabove5yrs["operation_years"]))#Mean value of "operation_years" who are survive more than 5years
print("Mean of operation years who are dead",np.mean(haberman_diewithin5yrs["operation_years"]))
print("Mean of Operation years ie:overall(including alive and dead)",np.mean(haberman["operation_years"]))
print("Mean of Lumph_nodes who are alive", np.mean(haberman_surviveabove5yrs["Lymph_nodes"]))
print("Mean of Lumph_nodes eho are dead",np.mean(haberman_diewithin5yrs["Lymph_nodes"]))
print("Mean of Operation yearsie:overall(including alive and dead)",np.mean(haberman["Lymph_nodes"]))
```

```python
print("standard deviation of Age who are alive",np.std(haberman_surviveabove5yrs["Age"]))
print("standard deviation of Age who are dead",np.std(haberman_diewithin5yrs["Age"]))
print("standard deviation of Age ie:overall(including alive and dead)",np.mean(haberman["Age"]))
print("standard deviation of operation years who are alive", np.std(haberman_surviveabove5yrs["operation_years"]))
print("standard deviation of operation years who are dead",np.std(haberman_diewithin5yrs["operation_years"]))
print("standard deviation of Operation years ie:overall(including alive and dead)",np.std(haberman["operation_years"]))
print("standard deviation of Lumph_nodes who are alive", np.std(haberman_surviveabove5yrs["Lymph_nodes"]))
print("standard deviation of Lumph_nodes eho are dead",np.std(haberman_diewithin5yrs["Lymph_nodes"]))
print("standard deviation of Operation yearsie:overall(including alive and dead)",np.std(haberman["Lymph_nodes"]))
```

```python
print("median of Age who are alive",np.median(haberman_surviveabove5yrs["Age"]))
print("median of Age who are dead",np.median(haberman_diewithin5yrs["Age"]))
print("median deviation of Age ie:overall(including alive and dead)",np.median(haberman["Age"]))
print("median of operation years who are alive", np.median(haberman_surviveabove5yrs["operation_years"]))
print("median of operation years who are dead",np.median(haberman_diewithin5yrs["operation_years"]))
print("median of Operation years ie:overall(including alive and dead)",np.median(haberman["operation_years"]))
print("median of Lumph_nodes who are alive", np.median(haberman_surviveabove5yrs["Lymph_nodes"]))
print("median of Lumph_nodes eho are dead",np.median(haberman_diewithin5yrs["Lymph_nodes"]))
print("median of Operation yearsie:overall(including alive and dead)",np.median(haberman["Lymph_nodes"]))
```

```python
print("Quantiles of Age who are alive ",np.percentile(haberman_surviveabove5yrs["Age"],np.arange(0,100,25)))
print("Quantiles of having lumph nodes who are alive",np.percentile(haberman_surviveabove5yrs["Lymph_nodes"],np.arange(0,100,25)))
print("Quantiles of after operation who are alive",np.percentile(haberman_surviveabove5yrs["operation_years"],np.arange(0,100,25)))
print("Quantiles of Age who are died ",np.percentile(haberman_diewithin5yrs["Age"],np.arange(0,100,25)))
print("Quantiles of having lumph nodes who are died",np.percentile(haberman_diewithin5yrs["Lymph_nodes"],np.arange(0,100,25)))
print("Quantiles of after operation who are died",np.percentile(haberman_diewithin5yrs["operation_years"],np.arange(0,100,25)))
print("Quantiles of Age who are alive(90%) ",np.percentile(haberman_surviveabove5yrs["Age"],90))
print("Quantiles of having lumph nodes who are alive(90%)",np.percentile(haberman_surviveabove5yrs["Lymph_nodes"],90))
print("Quantiles of after operation who are alive(90%)",np.percentile(haberman_surviveabove5yrs["operation_years"],90))
print("Quantiles of Age who are died(90%) ",np.percentile(haberman_diewithin5yrs["Age"],90))
print("Quantiles of having lumph nodes who are died(90%)",np.percentile(haberman_diewithin5yrs["Lymph_nodes"],90))
print("Quantiles of after operation who are died(90%)",np.percentile(haberman_diewithin5yrs["operation_years"],90))

#Quantiles that gives 100% value,
#25%values(Q1)["median of lower half of the dataset"],
#50%values(Q3)["median of sorted dataset"]
#75%values(Q2)["median of upper half of the data set"]
###And also we can give 90 to get 90% of result
```

```python
counts,bin_edges=np.histogram(haberman_surviveabove5yrs["Age"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
print(pdf)
print(cdf)
print(bin_edges)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.legend(["Age-pdf","Age-cdf"])
#1D plot
#Blue line is PDF,Orange line is CDF.
#And also I have printed the value of pdf,cdf,bin_edges
#In this plot x=bin_edges,y=pdf,cdf.
#At the age of 77 100% are survived
```

```python
counts,bin_edges=np.histogram(haberman_diewithin5yrs["Age"],bins=20,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
print(pdf)
print(bin_edges)
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title("histogram of age who are died")
plt.legend(["Age-pdf","Age_cdf"])
```

```python
counts,bin_edges=np.histogram(haberman_surviveabove5yrs["operation_years"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title("histogram of who are alive after operation of years")
plt.legend(["yearsofoperation-pdf","yearsofoperation_cdf"])
```
